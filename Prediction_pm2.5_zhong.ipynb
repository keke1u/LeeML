```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df1 = df[df['observation'] == 'PM2.5']
df1
```

```python
pm = df1.iloc[:,3:]
pm
```

```python
# split train data
# choose continuous 9 columns as feature which used to be trained
# the tenth as label
```

```python
# if we choose continuous 10 columns randomly, there will be  24-10+1=15 in one da
```

```python
train_x = []
train_y = []
for i in range(15):
    x = pm.iloc[:,i:i+9]
    x.columns = np.array(range(9)) # if we don't set columns, it will have different column names in each iteration
    y = pm.iloc[:,i+9]
    y.columns = np.array(range(1))
    train_x.append(x)
    train_y.append(y)
```

```python
train_x = pd.concat(train_x)
train_y = pd.concat(train_y)
# concat to a dataframe
train_x
```

```python
# change data type to array
train_x = np.array(train_x,float)
train_y = np.array(train_y,float)
train_x
```

```python
# batch gradient descent
w = np.zeros(train_x.shape[1])
b = np.ones(train_x.shape[0])
lr = 0.00000001
iteration = 10000
m = train_x.shape[0]
for i in range(iteration):
    yhat = np.dot(train_x, w) + b
    err = train_y - yhat
    dw = -np.dot(train_x.T, err)
    db = -np.sum(err)
    w = w - lr * dw
    b = b - lr * db
    if i%1000 ==0:
        print('after {} epochs, the loss on train data is:'.format(i),np.sum(err)/m)
print('after 10000 epochs, the loss on train data is:',np.sum(err)/m)
```

```python
# ada gradient descent
w = np.zeros(train_x.shape[1])
b = np.ones(train_x.shape[0])
lr = 10
iteration = 10000
m = train_x.shape[0]
sum_dw = np.zeros(train_x.shape[1])
sum_db = np.ones(train_x.shape[0])
for i in range(iteration):
    yhat = np.dot(train_x, w) + b
    err = train_y - yhat
    dw = -np.dot(train_x.T, err)
    db = -np.sum(err)
    sum_dw += dw**2
    sum_db += db**2
    ada_w = np.sqrt(sum_dw)
    ada_b = np.sqrt(sum_db)
    w = w - lr * dw / ada_w
    b = b - lr * db / ada_b
    if i%1000 ==0:
        print('after {} epochs, the loss on train data is:'.format(i),np.sum(err)/m)
print('after 10000 epochs, the loss on train data is:',np.sum(err)/m)
```
